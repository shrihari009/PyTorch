y = w * x + b

```python
x = np.random.randint(10,100,(6,3))
x
```

```python
# weights initialization
w = np.random.randn(1,3)
w
```

```python
# bias initilization
b = np.random.randn(6)
b
```

```python
x1 = torch.tensor(x)
w1 = torch.tensor(w, requires_grad=True)
b1 = torch.tensor(b, requires_grad=True)
```

```python
target = np.random.randint(10,100,6)
target
```

```python
target = torch.from_numpy(target)
```

```python
target
```

```python
def linear_model(x):
  return x.double() @ torch.t(w1) + b1
```

```python
x1.shape
```

```python
w1.shape
```

```python
pred = linear_model(x1)
```

```python
pred
```

```python
def mse(pred, target):
  diff = pred - target
  return torch.sum(diff*diff) / diff.numel()
```

```python
mse(pred, target)
```
